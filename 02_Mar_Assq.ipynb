```python
#Ans.2:-
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
```

```python
plt.scatter(x, y)

plt.title("First_Plot")
plt.xlabel("Number")
plt.ylabel("length")

plt.show()
```
